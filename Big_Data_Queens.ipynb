## MCC
El Merchant Category Code (MCC) es un código utilizado para identificar el tipo de mercado de una transacción. En el dataset, los mercados más solicitados son los siguientes:

- `5411`: Supermercados
- `5499`: Tiendas de alimentos/conveniencia y máquinas dispensadoras
- `5541`: Estaciones de servicio
- `5812`: Restaurantes
- `5912`: Farmacias
- `4829`: Transferencias bancarias
- `4784`: Peajes de carreteras y puentes
- `5300`: Tiendas mayoristas
- `4121`: Taxis y limusinas
